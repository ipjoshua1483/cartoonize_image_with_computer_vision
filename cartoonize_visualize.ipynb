# Cartoonize Images with Computer Vision

1. Upload image

```python
import cartoonize
cartoonize_obj = cartoonize.Cartoonize()
```

```python
cartoonize_obj.display_image()
```

2. Reduce the color palette via color quantization. This is done by using K-means clustering.

```python
# total_colors = 9
cartoonize_obj.color_quantization()
cartoonize_obj.display_image()
```

3. Apply a bilateral filter to further blur the image

```python
cartoonize_obj.bilateral_filter()
cartoonize_obj.display_image()
```

4. Create an edge mask and use bitwise operations to insert them into the image

```python
cartoonize_obj.edge_mask()
cartoonize_obj.display_image()
```
